```python
import numpy as np
import pandas as pd
```

```python
netflix = pd.read_csv("Copy of NETFLIX MOVIES and TV SHOWS CLUSTERING.CSV")
netflix.head(0)
```

```python
#show_id
#titlr
#type
#director
#country
#description
#release_year
#listed_in
```

```python
netflix=netflix[['show_id','title','type','country','listed_in','director','cast','description','release_year']]
```

```python
netflix.isnull().sum()
```

```python
netflix.dropna(inplace=True)
```

```python
netflix.head(8)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(x):
  y=[]
  for i in x.split():
   y.append(ps.stem(i))
  return " ".join(y)
```

```python
netflix['director']=netflix['director'].str.replace(" ","")
netflix['country']=netflix['country'].str.replace(" ","")
netflix['director']=netflix['director'].str.replace(" ","")
netflix['cast']=netflix['cast'].str.replace(" ","")
netflix['cast']=netflix['cast'].str.replace(","," ")
netflix['listed_in']=netflix['listed_in'].str.replace(" ","")
```

```python
netflix['tags'] = netflix['description']+netflix['listed_in']+' '+netflix['director']+' '+netflix['cast']
```

```python
nullflix=netflix[['show_id','title','tags']]
```

```python
nullflix['tags']=nullflix['tags'].apply(stem)
```

```python
nullflix['tags'][2]
```

```python
nullflix['tags'] = nullflix['tags'].apply(lambda x:x.lower())
```

```python
nullflix['tags'][2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(nullflix['tags']).toarray()
```

```python
vector[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=nullflix[nullflix['title']==movie].index[1]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(nullflix.iloc[i[1]].title)
    
```

```python
recommend("3 idiots")
```
